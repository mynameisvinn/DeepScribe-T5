```python
import os
import numpy as np
import pandas as pd
import sagemaker
from sagemaker.pytorch import PyTorch
```

```python
sagemaker_session = sagemaker.Session()
role = "SageMakerRole"
```

```python
estimator = PyTorch(
    entry_point="train.py",
    source_dir="scripts",
    role=role,
    framework_version="1.3.1",
    dependencies=['scripts/t1000'],
    py_version="py3",
    instance_count=1,
    instance_type='local',
)
```

```python
data_dir = os.path.join(os.getcwd(), 'data')
training = f'file://{data_dir}'
training
```

```python
inputs = {'training': training}
estimator.fit(inputs)
```
