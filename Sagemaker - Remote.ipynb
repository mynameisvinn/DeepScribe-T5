```python
import os
import numpy as np
import pandas as pd
import sagemaker
from sagemaker.pytorch import PyTorch
```

```python
sagemaker_session = sagemaker.Session()
role = "SageMakerRole"
```

```python
bucket = sagemaker_session.default_bucket()
prefix = "sagemaker/deepscribe"
training = sagemaker_session.upload_data('data', key_prefix=prefix)
training
```

```python
estimator = PyTorch(
    entry_point="train.py",
    source_dir="scripts",
    role=role,
    framework_version='1.5.0',
    dependencies=['scripts/t1000'],
    py_version="py3",
    instance_count=1,
    instance_type='ml.c4.xlarge',
    script_mode=True
)
```

```python
inputs = {'training': training}
estimator.fit(inputs)
```
